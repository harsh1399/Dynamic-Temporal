```python
with open("officeholders_entities.txt", "r") as f:
    text = f.read()
    titles = text.split("\n")
titles
```

```python
import json
with open(f"link_timeline.json") as f:
    links = json.load(f)
```

```python
len(links)
```

```python
links['Stade_Rennais_F.C.']
```

```python
k = links.keys()
```

```python
new_links = dict()
for t in titles:
    if t in k:
        new_links[t] = links[t]
```

```python
len(new_links.keys())
```

```python
with open(f"new_link_timeline.json", "w", encoding="utf-8") as f:
    json.dump(new_links, f, indent=4)
    f.close()
```
